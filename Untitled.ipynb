```python
import math
def f(k, x):
    ans = 0
    for i in range(k):
        ans += math.sin((2 * i + 1) * x) / (2 * i + 1)
    ans *= 4 / math.pi
    # ans -= base_line(x)
    return ans

def base_line(x):
    if x < 0:
        return -1
    return 1
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

```

```python
def plot_k(b, e):
    X = np.linspace(-math.pi, math.pi, 100)
    good_f = np.vectorize(f)
    good_base_line = np.vectorize(base_line)
    for i in range(b, e):
        Y = good_f(i, X)

        plt.plot(X, Y, linewidth = 1)

    plt.show()
```

```python
from matplotlib.pyplot import figure
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

# Major ticks every 20, minor ticks every 5
major_ticks = np.arange(-1.5, 1.5, 0.1)
minor_ticks = np.arange(-1.5, 1.5, 0.025)

ax.set_xticks(major_ticks)
ax.set_xticks(minor_ticks, minor=True)
ax.set_yticks(major_ticks)
ax.set_yticks(minor_ticks, minor=True)

# And a corresponding grid
ax.grid(which='both')
fig = plt.gcf()
fig.set_size_inches(10, 10)

plot_k(320, 325)
```
