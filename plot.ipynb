```python
items_list = items.drop_duplicates(subset=['item_id'])[['item_id', 'item_name', 'smth1', 'item_category_id', 'item_category_name', 'item_category_id_addon', 'item_category_name_addon']]                    
print(items_list.shape)
```

```python
from dateutil.rrule import rrule, DAILY
from datetime import datetime
start_date = datetime(2015, 1, 3)
end_date = datetime(2018, 1, 3)
dates = list(rrule(freq=DAILY, dtstart=start_date, until=end_date))
str_dates = [date.strftime('%Y-%m-%d') for date in dates]
```

```python
list(items_list.iloc[1])
```

```python
def combine(items_list_, dates_):
    list_of_lists = []
    for i in range(len(dates_)):
        for j in range(items_list_.shape[0]):
            A = list(items_list.iloc[j]) + [0, 0, 0, 0, 0, dates_[i], 0,0]
            list_of_lists.append(A)
    return pandas.DataFrame(list_of_lists, columns=['item_id', 'item_name', 'smth1', 'item_category_id', 'item_category_name', 'item_category_id_addon', 'item_category_name_addon', 'action1', 'action2', 'action3', 'action4', 'return', 
     'date', 'item_cnt_day', 'income'])

zeroes = combine(items_list, str_dates)
zeroes.to_csv('fill.csv')
```

```python
items_with_zeroes = pandas.concat([items, zeroes])
items_with_zeroes.to_csv('result.csv')
```

```python
print(items_with_zeroes.shape[0], items.shape[0] + zeroes.shape[0])
```
