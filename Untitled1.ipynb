```python
def euler_point(x0, t0, t_end, step, f):
    x = x0
    t = t0
    while t < t_end:
        x += step * f(t, x)
        t += step
    return x + (t_end - t) * f(t, x)

def steps(x0, t0, t_end, f, sol, eps):
    step = 2
    real_value = sol(t_end)
    while True:
        value = euler_point(x0, t0, t_end, (t_end - t0) / step, f)
        print(np.abs(value-real_value))
        if np.abs(value - real_value) < eps:
            return step
        step += 1
```
